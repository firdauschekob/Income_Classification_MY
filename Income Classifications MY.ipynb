```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn 
%matplotlib inline
```

```python
income_df = pd.read_csv("C:\\Users\\hp\\Documents\\Salaries & Wages\\salaries_certification_sex.csv")
income_df.head()
```

```python
income_df.describe()
```

```python
income_df.info()
```

```python
print('education',income_df.variable.unique())
```

```python
print('education',income_df.variable_en.unique())
```

```python
income_df[['variable_en', 'mean']].groupby(['variable_en'], as_index=False).mean().sort_values(by='mean', ascending=False)
```

```python
# Plot the distribution of mean wage by education
sns.boxplot(x="variable_en", y="mean", data=income_df, width=0.5)
plt.xticks(rotation=60)
plt.title("Mean Wage Distribution by Education")
plt.show()
```

```python
# Filter the data to only exclude 'overall' variable in 'sex'
filtered_data = income_df[income_df['sex'] != 'overall']

# Plot the distribution of mean wage by gender
sns.violinplot(x="sex", y="mean", data=filtered_data)
plt.title("Mean Wage Distribution by Gender")
plt.show()
```

```python
# Plot the relationship between year and mean wage
sns.lineplot(x="year", y="mean", data=income_df)
plt.title("Relationship between Year and Mean Wage")
plt.show()
```

```python
# Plot the distribution of mean wage by education and gender
sns.swarmplot(x="variable_en", y="mean", hue="sex", data=filtered_data)
plt.xticks(rotation=60)
plt.title("Mean Wage Distribution by Education and Gender")
plt.show()
```
